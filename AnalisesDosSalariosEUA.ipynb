```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando os dados em um DataFrame
salariesData = pd.read_csv('salarios.csv')
```

```python
salariesData
```

```python
salariesData.dropna(inplace=True)
```

```python
salariesData.to_csv('salarios_limpos2.csv', index=False)
```

```python
df2 = pd.read_csv('salarios_limpos2.csv')
```

```python
df2
```

```python
df2['Employee Annual Salary'].fillna(0)
```

```python
df2.dtypes
```

```python
df2['Employee Annual Salary'] = df2['Employee Annual Salary'].str.replace('[^\d.]', '', regex=True)
```

```python
df2
```

```python
pd.to_numeric(df2['Employee Annual Salary'])
```

```python
df2['Employee Annual Salary'] = pd.to_numeric(df2['Employee Annual Salary'], errors='coerce')
```

```python
df2['Employee Annual Salary'].dtypes
```

```python
df2
```

```python
df2['Monthly Salary'] = df2['Employee Annual Salary'] / 12
```

```python
df2
```

```python
df2['Monthly Salary'] = df2['Monthly Salary'].round(1)
```

```python
df2
```

```python
df2['Monthly Salary'].dtypes
```

```python
media_por_posicao = df2.groupby('Position Title')['Monthly Salary'].mean()
```

```python
df_media_salarios = pd.DataFrame(media_por_posicao).reset_index()
```

```python
df_media_salarios
```

```python
df_media_salarios.sort_values(by='Monthly Salary', ascending=False, inplace=True)
```

```python
df_media_salarios
```

```python
plot20MaioresSalariosChicago = df_media_salarios.head(20) 
# 20 maiores salários de Chicago 
```

```python
plot20MaioresSalariosChicago
```

```python
plt.figure(figsize=(10, 6))  # Ajuste o tamanho conforme necessário
plt.barh(plot20MaioresSalariosChicago['Position Title'], plot20MaioresSalariosChicago['Monthly Salary'], color="green")
plt.xticks(rotation=0)  # Rotaciona os rótulos em 45 graus
plt.xlabel('Monthly Salary')
plt.ylabel('Position Title')
plt.title('Os 20 Maiores Médias Saláriais de Chicago')

plt.show()
```

```python
df_media_salarios
```

```python
df_media_salarios.sort_values(by='Monthly Salary', ascending=True, inplace=True)
```

```python
df_media_salarios
```
